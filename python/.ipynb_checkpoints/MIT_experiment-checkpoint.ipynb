## Load data

```python
filename = '../data/mit.rds'
readRDS = robjects.r['readRDS']
res = readRDS(filename)
df = np.array(res).astype(int)
```

```python
df.shape
```

```python
df[0,:10,:10]
```

```python
num_nodes = df[0].shape[0]
# n = 20
T = len(df)
# num_changepts = 5

form_terms = ['edges', 'mutual']#, 'mutual'] #, 'mutual', 'triangle']
diss_terms = ['edges', 'mutual']# , 'mutual'] #, 'mutual', 'triangle']

p_pos = len(form_terms) # edge + mutual
p_neg = len(diss_terms) # edge + mutual
p = p_pos + p_neg


```

```python
from mple_learn import stergmGraph
```

```python
sim1 = stergmGraph(
    X = df,
    form_terms=form_terms,
    diss_terms=diss_terms,
    lam=[10],
    max_steps=100,
    newton_max_steps=100,
    admm_alpha=1,
    gfl_tol=1e-4,
#     gfl_maxit=200,
    converge_tol=1e-5,
    verbose=1
)
```

```python
np.random.seed(42)
theta = np.random.normal(0, 0.2, size= (T, p)) 
z = np.copy(theta) 
u = np.zeros((T, p)) 

init = {
    'theta': np.copy(theta),
    'z': z,
    'u': u
}
```

```python
start = timeit.default_timer()

theta_res, z, u, c = sim1.mple(initial_values=init, solver='newton', tau_inc=2, m=10)

stop = timeit.default_timer()
print('Time: ', stop - start)  
```
